```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
m = np.linspace(40, 200)
g=9.81; cd=0.25; t=4; v=36
fm=np.sqrt(g*m/cd)*np.tanh(np.sqrt(g*cd/m)*t)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(m, fm, 'yo-')
plt.title('Problem of significant vertebrae injury')
plt.xlabel('Mass')
plt.ylabel('Velocity')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.title('A tale of 2 subplots')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
ARCTAN=np.arctan(X)
plt.title("sin, cos, arctan 그래프")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.plot(X, ARCTAN, ls="-.", label="arctane")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.legend(loc=2)
plt.show()
```
